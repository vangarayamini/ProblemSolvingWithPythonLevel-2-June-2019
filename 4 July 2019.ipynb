```python
# Velocity = acceleration * time if a is constant v is directly proportional to time --> linearly proportional
# If a is not constant it is non linearly proportional

# V = a*t

t=list(range(1,11))
#a=list(range(len(t)+1))
v=[2*i for i in range(1,11)]
#v

plt.plot(t,v)

plt.grid()

plt.show()
```

```python
t=list(range(1,11))
a=list(range(1,len(t)+1))
v=[a[i]*i for i in range(len(t))]
#v

plt.plot(t,v)

plt.grid()

plt.show()
```

```python
n=list(range(10))
sqn=[i**2 for i in n]

plt.plot(n,sqn)
plt.show()
```

```python
# Voltage - Time
import numpy as np
t=[i/10 for i in range(1,21)]
#t
#t=np.arange(0.1,2.0,0.1)
v=[1+np.sin(2*np.pi*i) for i in t]
plt.title('Time vs Voltage Graph')
plt.plot(t,v)
plt.grid()
plt.xlabel('Time')
plt.ylabel('Voltage')
plt.show()
```
